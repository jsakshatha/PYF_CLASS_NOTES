# ML Pipeline with Categorical and Numerical Data

## 1. Import Libraries

```python

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

## 2. Create Dataset

```python

data = {
    "Age": [22, 25, 47, 52, 46, 56, 23, 40],
    "Salary": [15000, 29000, 48000, 52000, 46000, 58000, 18000, 41000],
    "Gender": ["Male", "Female", "Female", "Male", "Male", "Female", "Female", "Male"],
    "Purchased": [0, 0, 1, 1, 1, 1, 0, 1]
}

df = pd.DataFrame(data)
df
```

## 3. Split Features and Target

```python

X = df.drop("Purchased", axis=1)
y = df["Purchased"]
```

## 4. Train-Test Split

```python

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)
```

## 5. Preprocessing

```python

num_features = ["Age", "Salary"]
cat_features = ["Gender"]

preprocessor = ColumnTransformer(transformers=[
    ("num", StandardScaler(), num_features),
    ("cat", OneHotEncoder(drop="first"), cat_features)
])
```

## 6. Build Pipeline

```python

pipeline = Pipeline(steps=[
    ("preprocessing", preprocessor),
    ("model", LogisticRegression())
])
```

## 7. Train Model

```python

pipeline.fit(X_train, y_train)
```

## 8. Predictions

```python

y_pred = pipeline.predict(X_test)
y_pred
```

## 9. Evaluation

```python

print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```
